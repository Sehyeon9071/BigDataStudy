```python
# if문은 왜 필요한가

"돈이 있으면 택시를 타고 가고, 돈이 없으면 걸어간다."
```

```python
money = True

while True :
    try :
        money = int(input("현재 가진 돈을 입력하시오. : "))
        break

    except ValueError:
        print("숫자를 입력하시오.")

if money >= 5000: 
    print("택시를 타고 가라.")
    
else :
    print("걸어가라.")
```

```python
# if문의 기본 구조

# if 조건문 : 
#     수행할 문장1
#     수행할 문장2
#     ...

# else :
#     수행할 문장A
#     수행할 문장B
#     ...
```

```python
# 들여쓰기 방법

# if 조건문 :               
#     수행할 문장1
# 수행할 문장2      <-- 들여쓰기 오류 발생
#     수행할 문장3  

money = True
if money :
    print("택시를")
    print("타고")        
    print("가라.")        
```

```python
# 만약에 주머니에 money가 있으면 택시를 타고 가고,
# 그렇지 않으면 걸어가라.

pocket = ["paper", "phone", "money"]

if "money" in pocket :
    print("택시를 타고 가라.")

else :
    print("걸어가라.")
```

```python
# 다양한 조건을 판단하는 elif

pocket = ["paper", "phone"]
card = True

if "money" in pocket :
    print("택시를 타고 가라.")

else : 
    if card == True :
        print("택시를 타고 가라.")
    
    else :
        print("걸어가라.")
```

```python
pocket = ["paper", "phone"]
card = False

if "money" in pocket :
    print("택시를 타고 가라.")

elif card == True :
    print("택시를 타고 가라.")

else :
    print("걸어가라.")
```

```python
# 조건부 표현식

score = 85

if score >= 60 :
    result = "합격"

else :
    result = "불합격"

print(result)
```

```python
# 학점이 90 이상이면 A
# 80 이상 90 미만이면 B
# 70 이상 80 미만이면 C
# 60 이상 70 미만이면 D
# 60 미만이면 F

score = None

while True :
    try :
        score = int(input("점수를 입력하시오. : "))
        break

    except ValueError:
        print("숫자를 입력하시오.")

if score >= 90 :
    grade = "A"

elif score >= 80 :
    grade = "B"

elif score >= 70 :
    grade = "C"

elif score >= 60 :
    grade = "D"

else :
    grade = "F"

print(f"당신의 점수는 {score}점, 학점은 {grade}입니다.")
```
