```python
# 함수란 무엇인가
## 믹서기, 자판기

# 파이썬 함수의 구조

# def 함수이름(매개변수) :
#     수행할 문장1
#     수행할 문장2
#     ...

def add(a, b) :
    return a + b

result = add(3, 4)
print(result)
```

```python
# 일반적인 함수
## 입력값 : O, 반환값 : O

# def 함수이름(매개변수) :    - 매개변수 = 입력값
#     수행할 문장1
#     수행할 문장2
#     ...
#     return 결과값    - 결과값 = 반환값

# 예제
def add(a, b) :
    result = a + b
    return result

a = add(3, 4)
print(a)
```

```python
# 입력값이 없는 함수
## 입력값 : X, 반환값 : O

def say() :
    return "Hi"

a = say()
print(a)
```

```python
# 반환값이 없는 함수
## 입력값 : O, 반환값 : X

def add(a, b) :
    print(f"{a}, {b}의 합은 {a + b}이다.")

a = add(3, 4)
print(a)
```

```python
# 입력값 : X, 반환값 : X

def say() :
    print("Hi")

a = say()
print(a)
```

```python
# 매개변수 지정하여 호출하기

def sub(a, b) :
    return a - b

result1 = sub(7, 3)
print(result1)

result2 = sub(b = 7, a = 3)
print(result2)
```

```python
# 입력값이 몇 개가 될지 모를 때는 어떻게 해야 할까?

# def 함수이름(*매개변수) :     - *(1개)
#     수행할 문장1
#     수행할 문장2
#     ...

def add_many(*args) :   # args = [1, 2, 3, 4, 5]
    result = 0
    for i in args :
        result += i
    return result

a = add_many(1, 2, 3, 4 ,5)
print(a)
```

```python
# 키워드 매개변수, kwargs
## 키워드값의 형태로 매개변수를 받을때  - 매개변수 앞에 **(2개)를 붙인다.

def print_kwargs(**kwargs) :
    print(kwargs)

print_kwargs(a = 1)
print_kwargs(name = "foo", age = 3)
```

```python
def create_profile(**profiles) :
    print("===== 프로필 정보 =====")
    for key, value in profiles.items() :
        print(f"{key} : {value}")

create_profile(이름 = "김철수", 나이 = 30, 직업 = "개발자", 취미 = "운동")
```

```python
# *args, **kwargs 혼용

def mixed_function(name, *args, **kwargs) :
    print(f"이름 : {name}")
    print(f"추가인수들 : {args}")
    print(f"키워드인수들 : {kwargs}")

mixed_function("홍길동", 1, 2, 3, 4, 5, age = 25, city = "서울")
```

```python
def add_and_mul(a, b) :
    return a + b, a * b

result = add_and_mul(3, 4)
print(result)
```

```python
def say_nick(nick) :
    if nick == "바보" :
        return
    print(f"나의 별명은 {nick}입니다.")

# say_nick("야호")
say_nick("바보")
```

```python
# 매개변수의 초기값을 미리 설정하기

def say_myself(name, age, man = True) :
    print(f"나의 이름은 {name}입니다.")
    print(f"나의 나이는 {age}살 입니다.")
    if man :
        print("남자입니다.")
    else : 
        print("여자입니다.")

say_myself("홍길동", 30)
say_myself("영희", 30, False)
```

```python
# lambda 예약어

# def 함수이름(매개변수) :
#     수행할 문장
#     return

# def add(a, b) :
#     return a + b

# lambda
# 함수이름 = lambda 매개변수1, 매개변수2, ... : 수행할 문장

add = lambda a, b : a + b

result = add(3, 4)
print(result)
```

```python
def add(a, b) :
    """
    함수의 독스트링(Docstring)
    두 숫자를 더하는 함수
    설명 : 다른 사람 또는 나를 위한 문서
    """



    return a + b

result = add(3, 4)
print(result)

print(add.__doc__)
```
