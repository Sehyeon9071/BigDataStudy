```python
# 파일객체 = open(파일이름, 파일열기모드)
## r - 읽기 모드 : 파일을 읽기만 할 때
## w - 쓰기 모드 : 파일의 내용을 쓸 때
## a - 추가 모드 : 파일의 마지막에 내용을 추가할 때

#-*- coding: utf-8 -*-

f = open("data2.txt", "w")
f.close()
```

```python
#-*- coding: utf-8 -*-

f = open("./data/새파일.txt", "w")

for i in range(1, 11) :
    data = f"{i}번째 줄입니다. \n"
    f.write(data)

f.close()
```

```python
#-*- coding: utf-8 -*-

# f = open("C:/Users/NT371B5A/Documents/bigdata/BigDataStudy/03_입출력/data/새파일.txt", "r")
f = open("data/새파일.txt", "r")
line = f.readline()
print(line)
f.close()
```

```python
#-*- coding: utf-8 -*-

f = open("data/새파일.txt", "r")
while True :
    line = f.readline()
    if not line : break
    print(line)

f.close()
```

```python
# readlines() 함수 이용하기

#-*- coding: utf-8 -*-

f = open("data/새파일.txt", "r")
lines = f.readlines()

for line in lines :
    print(line, end="")
f.close()

# print(type(lines))
# print(len(lines))
```

```python
# read 함수 사용하기

#-*- coding: utf-8 -*-

f = open("data/새파일.txt", "r")
data = f.read()
print(data)
f.close()
```

```python
# 파일에 새로운 내용 추가하기

#-*- coding: utf-8 -*-

f = open("../새파일.txt", "a")

for i in range(11, 21) :
    f.write(f"{i}번째 줄입니다.")

f.close()
```

```python
# with 문과 함께 사용하기

#-*- coding: utf-8 -*-

with open("data3.txt", "w") as f :
    f.write("Life is short, you need Python")

```
