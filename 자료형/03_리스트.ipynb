```python
# 리스트의 길이 구하기

a = [1, 2, 3, 4, 5]
len(a)
```

```python
# 리스트의 수정과 삭제

a = [1, 2, 3]
a[2] = 4
a
```

```python
# del로 리스트 요소 삭제

a = [1, 2, 3]
del a[1]
print(a)
```

```python
a = [1, 2, 3, 4, 5]
del a[2:]
print(a)
```

```python
# 리스트 관련 함수

# .append()  - 리스트 요소 추가
a = [1, 2, 3]
a.append(4)
a
```

```python
# 리스트 정렬

# .sort() - 리스트 오름차순 정렬
a = [1, 4, 3, 2]
a.sort()
print(a)

b = ["a", "c", "d", "b"]
b.sort()
print(b)

# .sort(reverse = True) - 리스트 내림차순 정렬
c = [1, 4, 3, 2]
c.sort(reverse=True)
print(c)

d = ["a", "c", "d", "b"]
d.sort(reverse=True)
print(d)

# .reverse() - 리스트 반전
e = ["a", "c", "b"]
e.reverse()
print(e)
```

```python
# 리스트의 인덱스 반환

# .index()
a = [1, 2, 3]
print(a.index(1))
print(a.index(2))
print(a.index(3))
```

```python
# 리스트에 요소 삽입

# .insert(index, int)
a = [1, 2, 3]
a.insert(0, 4)
print(a)

a.insert(3, 5)
print(a)
```

```python
# 리스트 요소 제거

# .remove()
a = [1, 2, 3, 1, 2, 3]
a.remove(3)
print(a)
```

```python
# 리스트 요소 꺼내기

# .pop()
a = [1, 2, 3]
a.pop(1)
print(a)
```

```python
# 리스트에 포함된 요소 x의 개수 세기

# .count()
a = [1, 2, 3, 1]
a.count(1)
```

```python
# 리스트 확장

# .extend()
a = [1, 2, 3]
a.extend([4, 5])
print(a)
```
