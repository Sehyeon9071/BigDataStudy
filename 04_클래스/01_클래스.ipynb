```python
result1 = 0

def add1(num) :
    global result1
    result1 += num
    return result1

print(add1(3))
print(add1(4))
print(add1(5))
```

```python
class Calculator :
    def __init__(self) :
        self.result = 0
    
    def add(self, num) :
        self.result += num
        return self.result
    
cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(3))
print(cal1.add(4))

print(cal2.add(5))
print(cal2.add(10))
```

```python
# 곱하기, 빼기, 나누기 기능 만들기
## mul() - 곱하기
## sub() - 빼기
## div() - 나누기

class Calculator() :
    def setdata(self, first, second) :
        self.first = first
        self.second = second

    def mul(self) :
        result = self.first * self.second
        return result
    
    def sub(self) :
        result = self.first - self.second
        return result
    
    def div(self) :
        result = self.first / self.second
        return result
    
cal1 = Calculator()
cal2 = Calculator()
cal1.setdata(3, 4)
cal2.setdata(5, 6)

print(cal1.mul())
print(cal2.mul())
print(cal1.sub())
print(cal2.sub())
print(cal1.div())
print(cal2.div())
```

```python
class Calculator() :
    def setdata(self, first, second, ops) :
        self.first = first
        self.second = second
        self.ops = ops

    def cal(self) :
        dict_ops = {
            "add" : self.first + self.second,
            "sub" : self.first - self.second,
            "mul" : self.first * self.second,
            "div" : self.first / self.second
        }

        return dict_ops.get(self.ops, "올바르지 않은 연산자입니다.")
    

result = Calculator()
result.setdata(3, 4, "div")
print(result.cal())
```

```python
# 클래스의 상속
class MoreFourCal(Calculator) :
    pass

cal3 = MoreFourCal()
cal3.setdata(1, 2, "add")
cal3.cal()
```

```python
class MoreFourCal(Calculator) :
    def pow(self) :
        result = self.first ** self.second
        return result

cal4 = MoreFourCal()
cal4.setdata(2, 4, "add")
cal4.pow()
```

```python
# 매서드 오버라이딩

class SafeFourCal(Calculator) :
    def cal(self) :
        if self.second == 0 :
            return "0은 나눌 수 없습니다"
        else :
            dict_ops = {
                "add" : self.first + self.second,
                "sub" : self.first - self.second,
                "mul" : self.first * self.second,
                "div" : self.first / self.second
            }

            return dict_ops.get(self.ops, "올바르지 않은 연산자입니다.")

cal5 = SafeFourCal()
cal5.setdata(4, 0, "div")
cal5.cal()
```
