```python
# 역슬래시를 이용해서 작은따옴표와 큰따옴표를 문자열에 포함
# \ = 따옴표 문자로 인식, \n = 줄바꿈, \t = Tap 간격, \\ = \를 문자로 인식

food = "Python\'s favorite \n is \tperl"

print(food)
```

```python
# 문자열 인덱싱

a = "Life is too short"

print(a[8])
print(a[-1])
print(a[-2])
```

```python
# 문자열 슬라이싱

a = '''Life is too short, you need Python'''

print(a[0:4])
print(a[12:17])
print(a[5:])        # 끝까지
print(a[:17])       # 시작점부터
print(a[:])         # 전체
```
