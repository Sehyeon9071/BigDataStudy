```python
# 정수형

a = 123     # 양수
b = -178    # 음수
c = 0
print(a)
print(b)
print(c)
type(a)
```

```python
# 실수형

a = 1.2
b = -3.45
print(a)
print(b)
type(a)
```

```python
# 8진수와 16진수

a = 0o123
print(a)

b = 0x111
print(b)

0xF
```

```python
# 사칙연산

a = 3
b = 4

print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

```python
# x의 y제곱을 나타내는 ** 연산자

a = 3
b = 4

print(a ** b)
```

```python
# 나눗셈 후 나머지를 구하는 % 연산자

a = 7
b = 3
c = 2

print(a % b)
print(b % c)
```

```python
# 나눗셈 후 몫을 구하는 // 연산자

a = 7
b = 4

print(a // b)
```

```python
# 복합 연산자 +=, -=, *=, /=

a = 1
a += 1      # a = a + 1

print(a)
```
