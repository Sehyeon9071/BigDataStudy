```python
# 딕셔너리 만드는 방법

dic = {
        "name" : "park",
        "phone" : "010-1234-1234",
        "birth" : "00-01-01"
    }
```

```python
# 딕셔너리 요소 추가하기

dic["hobby"] = "soccer"

print(dic)
```

```python
dic[3] = [1, 2, 3]

print(dic)
```

```python
# 딕셔너리 요소 삭제

del dic[3]

print(dic)
```

```python
del dic["hobby"]

print(dic)
```

```python
# 딕셔너리 요소 변경하기

dic["name"] = "python"

print(dic) 
```

```python
# 딕셔너리 Key를 사용하여 Value 얻기

dic["name"]

print(dic["name"])
```

```python
print(dic["name"])
print(dic["phone"])
print(dic["birth"])
```

```python
# 딕셔너리 관련 함수

# key 리스트 만들기 - keys() = 딕셔너리의 key값만 리스트로 반환
dic = {
        "name" : "park",
        "phone" : "010-1234-1234",
        "birth" : "00-01-01"
    }

print(dic.keys())
print(list(dic.keys()))
```

```python
# Value 리스트 만들기 - values() = 딕셔너리의 value값만 리스트로 반환
print(dic.values())
print(list(dic.values()))
```

```python
# key, value 쌍 얻기 - items() = 딕셔너리의 key와 value를 튜플로 묶어서 리스트로 반환
print(dic.items())
print(list(dic.items()))
```

```python
# Key, value 쌍 모두 지우기 - clear() = 딕셔너리의 모든 요소 삭제
dic.clear()
print(dic)
```

```python
dic = {
        "name" : "park",
        "phone" : "010-1234-1234",
        "birth" : "00-01-01"
    }
```

```python
# Key로 Value 얻기 - get() = 딕셔너리의 key로 value값 얻기

print(dic.get("name"))
print(dic.get("phone"))
print(dic.get("birth"))
```

```python
# 해당 Key가 딕셔너리 안에 있는지 조사 - in = 딕셔너리 안에 해당 key가 있는지 조사하여 True/False 반환

print("name" in dic)
print("email" in dic)
```

```python
dic = {
        "name" : "park",
        "phone" : "010-1234-1234",
        "birth" : "00-01-01"
    }
```

```python
# Key로 Value 얻기 - pop() = 딕셔너리의 key로 value값 얻은 후 해당 key, value 쌍 삭제

phone = dic.pop("phone")
print(phone)
print(dic)
```
